```python
import colorcet as cc
import matplotlib.pyplot as plt
import numpy as np
```

```python
font1 = 28
font2 = 24
font3 = 18
font4 = 16
```

```python
num_el = 101
num_az = 2*101
itd = np.fromfile('data/itd_grid_eik.bin').reshape(num_el, num_az)

El = np.pi/2 + np.arcsin(np.linspace(-1, 1, num_el))
Az = np.linspace(0, 2*np.pi, num_az)

plt.figure(figsize=(13, 6))
plt.contourf(Az, El, 1e3*itd, np.linspace(-0.8, 0.8, 17), cmap=cc.cm.coolwarm)
clb = clb = plt.colorbar(format='%.1f ms')
clb.ax.tick_params(labelsize=font4)
clb.ax.tick_params(labelsize=font4)
plt.title('Eikonal ITD', fontsize=font1)
plt.xlabel(r'$\varphi$ (Azimuth)', fontsize=font2)
plt.ylabel(r'$\theta$ (Elevation)', fontsize=font2)
plt.yticks(np.linspace(0, np.pi, 5))
plt.gca().set_yticklabels([r'$0$', r'$\pi/4$', r'$\pi/2$', r'$3\pi/4$', r'$\pi$'], fontsize=font3)
plt.xticks(np.linspace(0, 2*np.pi, 5))
plt.gca().set_xticklabels([r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'], fontsize=font3)
plt.gca().set_aspect('equal')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
itd_bem_inf = np.fromfile('data/itd_grid_bem_inf.bin').reshape(num_el, num_az)
itd_bem_inf = np.where(np.isnan(itd), np.nan, itd_bem_inf)

plt.figure(figsize=(13, 6))
plt.contourf(Az, El, 1e3*itd_bem_inf, np.linspace(-0.8, 0.8, 17), cmap=cc.cm.coolwarm)
clb = clb = plt.colorbar(format='%.1f ms')
clb.ax.tick_params(labelsize=font4)
clb.ax.tick_params(labelsize=font4)
plt.title('BEM ITD at infinite source distance', fontsize=font1)
plt.xlabel(r'$\varphi$ (Azimuth)', fontsize=font2)
plt.ylabel(r'$\theta$ (Elevation)', fontsize=font2)
plt.yticks(np.linspace(0, np.pi, 5))
plt.gca().set_yticklabels([r'$0$', r'$\pi/4$', r'$\pi/2$', r'$3\pi/4$', r'$\pi$'], fontsize=font3)
plt.xticks(np.linspace(0, 2*np.pi, 5))
plt.gca().set_xticklabels([r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'], fontsize=font3)
plt.gca().set_aspect('equal')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
itd_bem = np.fromfile('data/itd_grid_bem.bin').reshape(num_el, num_az)
itd_bem = np.where(np.isnan(itd), np.nan, itd_bem)

plt.figure(figsize=(13, 6))
plt.contourf(Az, El, 1e3*itd_bem, np.linspace(-0.8, 0.8, 17), cmap=cc.cm.coolwarm)
clb = plt.colorbar(format='%.1f ms')
clb.ax.tick_params(labelsize=font4)
plt.title('BEM ITD at 600 mm source dist', fontsize=font1)
plt.xlabel(r'$\varphi$ (Azimuth)', fontsize=font2)
plt.ylabel(r'$\theta$ (Elevation)', fontsize=font2)
plt.yticks(np.linspace(0, np.pi, 5))
plt.gca().set_yticklabels([r'$0$', r'$\pi/4$', r'$\pi/2$', r'$3\pi/4$', r'$\pi$'], fontsize=font3)
plt.xticks(np.linspace(0, 2*np.pi, 5))
plt.gca().set_xticklabels([r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'], fontsize=font3)
plt.gca().set_aspect('equal')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(13, 6))
plt.contourf(Az, El, 1e3*itd, np.linspace(-0.7, 0.7, 51), cmap=cc.cm.coolwarm)
clb = plt.colorbar(format='%.1f ms')
clb.ax.tick_params(labelsize=font4)
plt.title('Eikonal-based ITD', fontsize=font1)
plt.xlabel(r'$\varphi$ (Azimuth)', fontsize=font2)
plt.ylabel(r'$\theta$ (Elevation)', fontsize=font2)
plt.yticks(np.linspace(0, np.pi, 5))
plt.gca().set_yticklabels([r'$0$', r'$\pi/4$', r'$\pi/2$', r'$3\pi/4$', r'$\pi$'], fontsize=font3)
plt.xticks(np.linspace(0, 2*np.pi, 5))
plt.gca().set_xticklabels([r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'], fontsize=font3)
plt.gca().set_aspect('equal')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(13, 6))
plt.contourf(Az, El, 1e3*itd_bem, np.linspace(-0.7, 0.7, 51), cmap=cc.cm.coolwarm)
clb = plt.colorbar(format='%.1f ms')
clb.ax.tick_params(labelsize=font4)
plt.title('BEM-based ITD', fontsize=font1)
plt.xlabel(r'$\varphi$ (Azimuth)', fontsize=font2)
plt.ylabel(r'$\theta$ (Elevation)', fontsize=font2)
plt.yticks(np.linspace(0, np.pi, 5))
plt.gca().set_yticklabels([r'$0$', r'$\pi/4$', r'$\pi/2$', r'$3\pi/4$', r'$\pi$'], fontsize=font3)
plt.xticks(np.linspace(0, 2*np.pi, 5))
plt.gca().set_xticklabels([r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'], fontsize=font3)
plt.gca().set_aspect('equal')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
diff_inf = abs(itd) - abs(itd_bem_inf)

plt.figure(figsize=(13, 6))
plt.contourf(Az, El, 1e3*diff_inf, np.linspace(-0.067, 0.067, 51), cmap=cc.cm.coolwarm)
clb = plt.colorbar(format='%.2f ms')
clb.ax.tick_params(labelsize=font4)
plt.title('Absolute error (Eikonal - BEM_inf_dist)', fontsize=font1)
plt.xlabel(r'$\varphi$ (Azimuth)', fontsize=font2)
plt.ylabel(r'$\theta$ (Elevation)', fontsize=font2)
plt.yticks(np.linspace(0, np.pi, 5))
plt.gca().set_yticklabels([r'$0$', r'$\pi/4$', r'$\pi/2$', r'$3\pi/4$', r'$\pi$'], fontsize=font3)
plt.xticks(np.linspace(0, 2*np.pi, 5))
plt.gca().set_xticklabels([r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'], fontsize=font3)
plt.gca().set_aspect('equal')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
diff = abs(itd) - abs(itd_bem)
max = np.nanmax(abs(diff))

plt.figure(figsize=(13, 6))
plt.contourf(Az, El, 1e3*diff, np.linspace(-max*1e3, max*1e3, 51), cmap=cc.cm.coolwarm)
clb = plt.colorbar(format='%.2f ms')
clb.ax.tick_params(labelsize=font4)
plt.title('Absolute error (Eikonal - BEM)', fontsize=font1)
plt.xlabel(r'$\varphi$ (Azimuth)', fontsize=font2)
plt.ylabel(r'$\theta$ (Elevation)', fontsize=font2)
plt.yticks(np.linspace(0, np.pi, 5))
plt.gca().set_yticklabels([r'$0$', r'$\pi/4$', r'$\pi/2$', r'$3\pi/4$', r'$\pi$'], fontsize=font3)
plt.xticks(np.linspace(0, 2*np.pi, 5))
plt.gca().set_xticklabels([r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'], fontsize=font3)
plt.gca().set_aspect('equal')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
fig = plt.figure(figsize=(13, 6))
ax = fig.add_subplot(111)
ax.hist(np.where(np.isnan(diff), 0, diff*1e3))
plt.title('Absolute error histogram', fontsize=font1)
plt.ylabel("Count")
plt.xlabel("Error (ms)")
```

```python
# Compute the relative error in the zone of interest where the ITD is above 0.025 ms
rel_error = np.where(abs(itd_bem) > 0.000025, diff / abs(itd_bem), 0)
rel_error = np.where(np.isnan(itd_bem), np.nan, rel_error)

plt.figure(figsize=(13, 6))
plt.contourf(Az, El, 100*rel_error, np.linspace(-34, 34, 51), cmap=cc.cm.coolwarm)
clb = plt.colorbar(format='%.0f%%')
clb.ax.tick_params(labelsize=font4)
plt.title('Relative error (Eikonal - BEM)', fontsize=font1)
plt.xlabel(r'$\varphi$ (Azimuth)', fontsize=font2)
plt.ylabel(r'$\theta$ (Elevation)', fontsize=font2)
plt.yticks(np.linspace(0, np.pi, 5))
plt.gca().set_yticklabels([r'$0$', r'$\pi/4$', r'$\pi/2$', r'$3\pi/4$', r'$\pi$'], fontsize=font3)
plt.xticks(np.linspace(0, 2*np.pi, 5))
plt.gca().set_xticklabels([r'$0$', r'$\pi/2$', r'$\pi$', r'$3\pi/2$', r'$2\pi$'], fontsize=font3)
plt.gca().set_aspect('equal')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.ticker as mtick

fig = plt.figure(figsize=(13, 6))
ax = fig.add_subplot(111)
ax.hist(np.where(np.isnan(rel_error), 0, rel_error*100))
ax.xaxis.set_major_formatter(mtick.PercentFormatter())
plt.title('Relative error histogram', fontsize=font1)
plt.ylabel("Count")
plt.xlabel("Error")
```
